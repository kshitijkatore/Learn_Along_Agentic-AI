` Matrix Multiplication `

* The inner dimensions must match:
*  (3, 2) @ (3, 2) won't work
*  (2, 3) @ (3, 2) will work
*  (3, 2) @ (2, 3) will work

````python
# Create a tensor and perform element-wise multiplication and matrix multiplication

import torch
tensor = torch.tensor([1,2,3,4])
tensor.shape
````

` Operation`	 `Calculation`	 `Code`
* Element-wise multiplication	[1*1, 2*2, 3*3] = [1, 4, 9]	`tensor * tensor`
* Matrix multiplication	`[1*1 + 2*2 + 3*3] = [14]`	`tensor.matmul(tensor)`


````python
# Element-wise matrix multiplication
tensor * tensor
````

````python
# Matrix multiplication
torch.matmul(tensor, tensor)
````

````python
# can also user "@" for matrix multiplication
tensor @ tensor
````

Matrix multiplication by hand

* The in-bulit torch.matmul() method is faster

````python
%%time

value =0

for i in range(len(tensor)):
    value += tensor[i] * tensor[i]
value
````

````python
%%time
torch.matmul(tensor, tensor)
````

````python
# Shape need to be in the right way

tensor_a = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])

tensor_b = torch.tensor([[7,10],
                        [8,11],
                        [9,12]])

torch.matmul(tensor_a,tensor_b)
````

`Transpose`

* torch.transpose(input, dim0, dim1) - where input is the desired tensor to transpose and dim0 and dim1 are the dimensions to be swapped.
* tensor.T - where tensor is the desired tensor to transpose.

````python
# view tensor_a and tensor_b
print(tensor_a)
print(tensor_b)
````

````python
# vies tensor_a and tensor_b.T

print(tensor_a)
print(tensor_b.T)
````

````python
# The operation works when tensor_b is transposed

output = torch.matmul(tensor_a, tensor_b.T)

print(output)
print(output.shape)
````

* You cal also use torch.mm() which is a short for torch.matmul().

````python
output = torch.mm(tensor_a, tensor_b.T)

print(output)
````
